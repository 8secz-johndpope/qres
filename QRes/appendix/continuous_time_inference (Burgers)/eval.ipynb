## Count Number of Parameters

```python
l = [2, 20, 20, 20, 20, 20, 20, 20, 20, 1]
num_param = 0
for i in range(len(l)-1):
    num_param += l[i] * l[i+1] + l[i+1]
print("PINN Num params: %.2fk" % np.round(num_param / 1000, 2))
```

```python
l = [2, 10, 10, 10, 10, 10, 10, 10, 10, 1]
num_param = 0
for i in range(len(l)-1):
    num_param += l[i] * l[i+1] * 2 + l[i+1]
print("QRes-lite Num params: %.2fk" % np.round(num_param / 1000, 2))
```

```python
l = [2, 14, 14, 14, 14, 14, 14, 14, 14, 1]
num_param = 0
for i in range(len(l)-1):
    num_param += l[i] * l[i+1] * 2 + l[i+1]
print("QRes-full Num params: %.2fk" % np.round(num_param / 1000, 2))
```

# Training Loss


```python
pinn = np.load("loss/loss_PINNs.npy")
qres = np.load("loss/loss_QRes.npy")
qres_adam = np.load("loss/loss_QRes_Adam.npy")
qres_adam_lite = np.load("loss/loss_QRes_Adam_lite.npy")
apinn = np.load("loss/loss_APINNs.npy")
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(3, 2.5))

x_apinn = np.arange(len(apinn)) * 10
x_qres_adam = np.arange(len(qres_adam)) * 10

sns.lineplot(data=pinn, label='PINN (L-BFGS-B)', alpha=0.8, lw=0.5, ax=ax)
sns.lineplot(data=qres, label='QRes (L-BFGS-B)', alpha=0.8, lw=0.5, ax=ax)
sns.lineplot(x=x_apinn, y=apinn, label='APINN (Adam)', alpha=0.8, lw=0.5, ax=ax)
sns.lineplot(x=x_qres_adam, y=qres_adam, label='QRes (Adam)', alpha=0.8, lw=0.5, ax=ax)
sns.lineplot(x=x_qres_adam, y=qres_adam_lite, label='QRes-lite (Adam)', alpha=0.8, lw=0.5, ax=ax)
ax.set_xlabel('Epoch')
ax.set_ylabel('Training Loss (log)')
ax.set_yscale('log')
ax.legend(loc=1)

l = ax.get_legend()
l.set_title('')
l.get_frame().set_linewidth(0.0)

sns.despine(offset=0)
fig.tight_layout()
fig.savefig("figures/loss_burgers_inference.pdf", dpi=300)
plt.show()
```
